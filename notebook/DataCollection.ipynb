```python
# API Key
API_KEY = MY_API_KEY
youtube = build('youtube', 'v3', developerKey=API_KEY)

# Masukkan ID Video YouTube
video_ids = [
    'cuL4Fl0RZII',
    'mz117p4_v2U',
    'dcE80aReHoU',
    'G7NfecMdGPE',
    'MoGAbf70wms',
    '8YITikwT9dM',
    'P9Q8J0UqKno',
    'DsOjjkPdP4A',
    '-TyJuaqOC-o',
    'BVeqk4An2IA',
    'TAvJFgfOi2Q',
    'iRcdUlXnnhY',
    'cjypb9NExf0',
    'dpBcR5viwCE',
    'O9W1wjzxyyI',
    'h_5w2olHow8',
    'QyHjzkF2uMs',
    'tQAkyWNuN-U',
    'n-Y5FhDL1Hk',
    '5FfbDtDiJ5A',
    'yZTbIcGRVbs',
    'Vq_NfyPdJjA'
]

all_comments = []

def get_replies(parent_id, video_id):
    """Fungsi khusus untuk menarik balasan dari sebuah komentar"""
    request = youtube.comments().list(
        part="snippet",
        parentId=parent_id,
        maxResults=100,
        textFormat="plainText"
    )

    while request:
        response = request.execute()
        for item in response['items']:
            reply = item['snippet']
            all_comments.append({
                'video_id': video_id,
                'comment_type': 'Reply',
                'author': reply['authorDisplayName'],
                'comment_text': reply['textDisplay'],
                'like_count': reply['likeCount'],
                'published_at': reply['publishedAt']
            })

        if 'nextPageToken' in response:
            request = youtube.comments().list(
                part="snippet",
                parentId=parent_id,
                pageToken=response['nextPageToken'],
                maxResults=100,
                textFormat="plainText"
            )
        else:
            break

def get_comments_from_video(video_id):
    print(f"Menarik komentar dari video: {video_id}...")
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=100,
        textFormat="plainText"
    )

    while request:
        response = request.execute()

        for item in response['items']:
            # Ekstrak komentar utama
            comment = item['snippet']['topLevelComment']['snippet']
            parent_id = item['snippet']['topLevelComment']['id']
            reply_count = item['snippet']['totalReplyCount']

            all_comments.append({
                'video_id': video_id,
                'comment_type': 'Top-Level',
                'author': comment['authorDisplayName'],
                'comment_text': comment['textDisplay'],
                'like_count': comment['likeCount'],
                'published_at': comment['publishedAt']
            })

            # Jika ada balasan, panggil fungsi get_replies
            if reply_count > 0:
                get_replies(parent_id, video_id)

        if 'nextPageToken' in response:
            request = youtube.commentThreads().list(
                part="snippet",
                videoId=video_id,
                pageToken=response['nextPageToken'],
                maxResults=100,
                textFormat="plainText"
            )
        else:
            break

# Eksekusi
for vid in video_ids:
    try:
        get_comments_from_video(vid)
    except Exception as e:
        print(f"Terjadi kesalahan pada video {vid}: {e}")

# Membuat dataframe
df = pd.DataFrame(all_comments)
print(f"Total {len(df)} komentar berhasil di ambil")
```
